```python
df=pd.read_csv("/content/insurance_data.csv")
```

```python
x=df[['age']]
print(type(x))
```

```python
y=df[['bought_insurance']]
print(type(y))
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
#training the model
reg.fit(x,y)
```

```python
#prediction
reg.predict([[45]])
```

```python
s=reg.score(x,y)
Accuracy=s*100
print("Accuracy is",Accuracy)
Loss=100,Accuracy
print("loss is",Loss)
```
